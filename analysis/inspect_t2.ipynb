```python
try:
    fig = inspector.plot_nobs(    selections=selections,
        labels=["Fiducial reduced", "Efficient reduced", "Fiducial oxidized", "Efficient oxidized"],
        colours=["#377eb8", "#e41a1c", "#4daf4a", "#ff7f00"],
        threshold=150,
        )

    fig.savefig(f"plots/nobs_{tier}.pdf", dpi=600)
except Exception as e:
    print(f"Error plotting nobs: {e}")
    pass


def analyze_nobs_by_category(inspector, selections=None):
    """
    Analyze the number of observations (nobs) required by planet category
    and identify which category requires the fewest observations on average.
    
    Parameters
    ----------
    inspector : Inspector object
        The inspector instance with loaded data
    selections : list of lists, optional
        List of indices for each planet type. If None, will create default selections.
    
    Returns
    -------
    dict : Analysis results with statistics for each category
    """
    import numpy as np
    
    # Create default selections if not provided
    if selections is None:
        selections = [
            list(np.where((inspector.out_targetlist["Migration Efficiency [Fiducial = 0, Efficient = 1]"]==0) & 
                         (inspector.out_targetlist["Redox [Reduced = 0, Oxidized = 1]"]==0))[0]),
            list(np.where((inspector.out_targetlist["Migration Efficiency [Fiducial = 0, Efficient = 1]"]==1) & 
                         (inspector.out_targetlist["Redox [Reduced = 0, Oxidized = 1]"]==0))[0]),
            list(np.where((inspector.out_targetlist["Migration Efficiency [Fiducial = 0, Efficient = 1]"]==0) & 
                         (inspector.out_targetlist["Redox [Reduced = 0, Oxidized = 1]"]==1))[0]),
            list(np.where((inspector.out_targetlist["Migration Efficiency [Fiducial = 0, Efficient = 1]"]==1) & 
                         (inspector.out_targetlist["Redox [Reduced = 0, Oxidized = 1]"]==1))[0])
        ]
    
    # Category labels
    category_labels = ["Fiducial reduced", "Efficient reduced", "Fiducial oxidized", "Efficient oxidized"]
    
    # Extract nobs data
    nobs_data = inspector.out_targetlist["nobs"]
    
    # Analyze each category
    results = {}
    category_stats = []
    
    print("=== Number of Observations Analysis by Category ===\n")
    
    for i, (selection, label) in enumerate(zip(selections, category_labels)):
        if len(selection) == 0:
            print(f"{label}: No planets in this category")
            continue
            
        category_nobs = nobs_data[selection]
        
        # Calculate statistics
        mean_nobs = np.mean(category_nobs)
        median_nobs = np.median(category_nobs)
        std_nobs = np.std(category_nobs)
        min_nobs = np.min(category_nobs)
        max_nobs = np.max(category_nobs)
        
        # Store results
        results[label] = {
            'mean': mean_nobs,
            'median': median_nobs,
            'std': std_nobs,
            'min': min_nobs,
            'max': max_nobs,
            'count': len(category_nobs),
            'selection_indices': selection
        }
        
        category_stats.append((label, mean_nobs))
        
        # Print statistics
        print(f"{label}:")
        print(f"  Count: {len(category_nobs)} planets")
        print(f"  Mean nobs: {mean_nobs:.2f}")
        print(f"  Median nobs: {median_nobs:.2f}")
        print(f"  Std deviation: {std_nobs:.2f}")
        print(f"  Range: {min_nobs} - {max_nobs}")
        print()
    
    # Find category with lowest mean nobs
    if category_stats:
        category_stats.sort(key=lambda x: x[1])  # Sort by mean nobs
        best_category = category_stats[0]
        
        print("=== Summary ===")
        print(f"Category requiring FEWEST observations on average:")
        print(f"  {best_category[0]}: {best_category[1]:.2f} observations")
        print()
        print("Ranking from lowest to highest mean observations:")
        for i, (cat, mean_obs) in enumerate(category_stats, 1):
            print(f"  {i}. {cat}: {mean_obs:.2f}")
        
        results['ranking'] = category_stats
        results['best_category'] = best_category
    
    return results

nobs_analysis = analyze_nobs_by_category(inspector, selections)
```

```python
def analyze_observation_budget(inspector, target_time, selections=None):
    """
    Analyze how many planets can be observed within a given time budget
    and their distribution by type.
    
    Parameters
    ----------
    inspector : Inspector object
        The inspector instance with loaded data
    target_time : float
        Target observation time budget
    selections : list of lists, optional
        List of indices for each planet type. If None, will create default selections.
    
    Returns
    -------
    dict : Analysis results including total planets, breakdown by type, etc.
    """
    from copy import deepcopy
    
    # Create default selections if not provided
    if selections is None:
        selections = [
            list(np.where((inspector.out_targetlist["Migration Efficiency [Fiducial = 0, Efficient = 1]"]==0) & 
                         (inspector.out_targetlist["Redox [Reduced = 0, Oxidized = 1]"]==0))[0]),
            list(np.where((inspector.out_targetlist["Migration Efficiency [Fiducial = 0, Efficient = 1]"]==1) & 
                         (inspector.out_targetlist["Redox [Reduced = 0, Oxidized = 1]"]==0))[0]),
            list(np.where((inspector.out_targetlist["Migration Efficiency [Fiducial = 0, Efficient = 1]"]==0) & 
                         (inspector.out_targetlist["Redox [Reduced = 0, Oxidized = 1]"]==1))[0]),
            list(np.where((inspector.out_targetlist["Migration Efficiency [Fiducial = 0, Efficient = 1]"]==1) & 
                         (inspector.out_targetlist["Redox [Reduced = 0, Oxidized = 1]"]==1))[0])
        ]
    
    # Type labels
    type_labels = ["Fiducial reduced", "Efficient reduced", "Fiducial oxidized", "Efficient oxidized"]
    
    # Get observation times and sort
    obs_times = deepcopy(inspector.out_targetlist["obs_time"].data)
    sorted_indices = np.argsort(obs_times)
    obs_times_sorted = obs_times[sorted_indices]
    
    # Calculate cumulative sum
    cum_time = np.cumsum(obs_times_sorted)
    
    # Find how many planets fit within the budget
    within_budget = cum_time <= target_time
    n_planets_total = np.sum(within_budget)
    
    if n_planets_total == 0:
        print(f"No planets can be observed within {target_time} time units")
        return {"total_planets": 0, "by_type": {}, "actual_time": 0}
    
    # Get indices of planets within budget
    planets_within_budget = sorted_indices[within_budget]
    actual_time_used = cum_time[n_planets_total - 1]
    
    # Count planets by type
    type_counts = {}
    for i, (selection, label) in enumerate(zip(selections, type_labels)):
        # Find intersection between planets within budget and this type
        planets_of_type = np.intersect1d(planets_within_budget, selection)
        type_counts[label] = len(planets_of_type)
    
    # Print results
    print(f"\n=== Observation Budget Analysis ===")
    print(f"Target time budget: {target_time}")
    print(f"Actual time used: {actual_time_used:.2f}")
    print(f"Total planets within budget: {n_planets_total}")
    print(f"\nBreakdown by type:")
    for label, count in type_counts.items():
        percentage = (count / n_planets_total) * 100 if n_planets_total > 0 else 0
        print(f"  {label}: {count} ({percentage:.1f}%)")
    
    # Return results dictionary
    return {
        "total_planets": n_planets_total,
        "actual_time": actual_time_used,
        "target_time": target_time,
        "by_type": type_counts,
        "planets_indices": planets_within_budget,
        "type_percentages": {label: (count / n_planets_total) * 100 if n_planets_total > 0 else 0 
                           for label, count in type_counts.items()}
    }

target_times = [250, 500, 1000]
results = {}

for target_time in target_times:
    results[target_time] = analyze_observation_budget(inspector, target_time, selections)
```
