```python
import numpy as np
import matplotlib.pyplot as plt

planet0 = np.loadtxt('out/ArielInstrumentTier2/Planet0/Planet0.txt').T
planet1 = np.loadtxt('out/ArielInstrumentTier2/Planet1/Planet1.txt').T
planet2 = np.loadtxt('out/ArielInstrumentTier2/Planet2/Planet2.txt').T

plt.errorbar(planet0[0], planet0[1], yerr=planet0[2], label='Planet 0')
plt.errorbar(planet1[0], planet1[1], yerr=planet1[2], label='Planet 1')
plt.errorbar(planet2[0], planet2[1], yerr=planet2[2], label='Planet 2')

plt.legend()
```
