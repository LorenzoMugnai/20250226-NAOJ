```python
   
import os
import numpy as np
from tqdm import tqdm

base_dir = "./20250322_ArielTargetProfileData_ChangedPnames"

keyword = "Compo"


files = [os.path.join(base_dir, name) for name in os.listdir(base_dir) if os.path.isfile(os.path.join(base_dir, name)) if name.endswith(".csv")]


for file_ in tqdm(files):
    skip_rows = 3 if keyword in file_ else 2
    
    try:
        data = np.loadtxt(file_, skiprows=skip_rows, delimiter=',')
        output_path = file_.replace(".csv", ".dat")
        np.savetxt(output_path, data)
    except Exception as e:
        print(f"Errore nel file {file_}: {e}")

```
