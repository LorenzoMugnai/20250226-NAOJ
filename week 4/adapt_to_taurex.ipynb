```python
   
import os
import numpy as np
from tqdm import tqdm

base_dir = "/home/ariel/lorenzo.mugnai/USER_DATA/Alfnoor2/20250226 - NAOJ/week 4/20250322_ArielTargetProfileData"

keyword = "Compo"

target_ext = ".csv"

for root, _, files in os.walk(base_dir):
    for file_ in files:
        if file_.endswith(target_ext):
            file_path = os.path.join(root, file_)

            skip_rows = 3 if keyword in file_ else 2
            
            try:
                data = np.loadtxt(file_path, skiprows=skip_rows, delimiter=',')
                output_path = file_path.replace(".csv", ".dat")
                np.savetxt(output_path, data)
            except Exception as e:
                print(f"Errore nel file {file_path}: {e}")

```
